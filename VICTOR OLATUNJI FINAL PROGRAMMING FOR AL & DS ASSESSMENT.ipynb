```python
import csv
Dependants_error =[] 
my_list = []
with open("acw_user_data.csv",mode="r") as file:
    my_acw_file = csv.reader(file,delimiter=",")
    for line_number,line in enumerate(my_acw_file):
        if line_number > 0:
            acw_user_data ={}
            Vehicle = {}  
            Credit_Card = {} 
            Address = {} 
            Address_Street = str(line[0])
            Address_City = str(line[1])
            Address_Postcode = str(line[2])
            Age= int(line[3])
            Distance_Commuted_to_work = float(line[4])
            Employer_Company = str(line[5])
            Credit_Card_Start_Date = str(line[6])
            Credit_Card_Expire_Date = str(line[7])
            Credit_Card_Number = int(line[8])
            Credit_Card_CVV = int(line[9])
            try:                             # exception method is used to avoid error in casting empty cell 
                Dependants = int(line[10])    
            except Exception as err:
                pass
            finally:                        # solution provided for the dependants column
                Dependants = line[10]
            if (Dependants == "" or Dependants==" "):
                Dependants_error.append(line_number+1) 
                Dependants = "NA"   ## replace void columns with N/A
            else:
                Dependants = int(line[10])
            First_Name = str(line[11])
            Bank_IBAN = str(line[12])
            Last_Name =str(line[13])
            Marital_Status =str(line[14])
            Yearly_Pension = float(line[15])
            Retired = str(line[16])
            if str(line[16]).lower() == "true":
                Retired = True
            else:
                Retired = False
            Yearly_Salary = float(line[17])
            Sex = str(line[18])
            Vehicle_Make = str(line[19])
            Vehicle_Model =str(line[20])
            Vehicle_Year = int(line[21])
            Vehicle_Type = str(line[22])

            Vehicle = {"Vehicle Make": Vehicle_Make, "Vehicle Model": Vehicle_Model, "Vehicle Year": Vehicle_Year,
                      "Vehicle Type": Vehicle_Type} 
            Credit_Card ={"Start Date": Credit_Card_Start_Date, "End Date": Credit_Card_Expire_Date, "Number":Credit_Card_Number, 
                         "Security Code": Credit_Card_CVV,"IBAN": Bank_IBAN }
            Address ={"Main Address": Address_Street, "City": Address_City, 
                      "Post Code": Address_Postcode} 
          
            acw_user_data={"First Name": First_Name,"Last Name": Last_Name,"Age": Age, "Sex": Sex, "Retired":Retired,
                          "Marital_Status": Marital_Status,"Dependants":Dependants,"Yearly Salary": Yearly_Salary,
                          "Yearly Pension":Yearly_Pension,"Employer Company": Employer_Company,"Distanct Commuted to work":
                            Distance_Commuted_to_work,"Vehicle":Vehicle,"Credit Card": Credit_Card,"Address": Address}
            my_list.append(acw_user_data)
print("Problematic rows for dependants:" ,Dependants_error)
print( )
print(my_list)
```

```python
import json
remove_ccard =[]
def fault_cc(my_list):   
    Start_month = int(my_list["Credit Card"]["Start Date"].split("/")[0])
    Start_year = int(my_list["Credit Card"]["Start Date"].split("/")[1])
    End_month = int(my_list["Credit Card"]["End Date"].split("/")[0])
    End_year = int(my_list["Credit Card"]["End Date"].split("/")[1])
    year_diff = End_year - Start_year
    if year_diff > 10:
        return True
    else:
        return False
    
for item in my_list:
    if fault_cc(item)==True:
        remove_ccard.append(item)

with open("remove_ccard.json", mode="w") as fault_cc:
    json.dump(remove_ccard,fault_cc)
    

    print("Total removed credit card: ",len(remove_ccard))
print( )
print(remove_ccard)
```

### TASK 7A

```python
with open("processed.json", mode="r") as processed_json:
    my_processed_file =json.load(processed_json)
commute = []
less_commute= []
more_commute = []
for items in my_processed_file:
    salary_earn = float(items["Yearly Salary"])
    distance_travel = float(items["Distanct Commuted to work"])
    if distance_travel <=1:
        items["Salary-Commute"]=salary_earn
        commute.append(items)
        less_commute.append(items)
    else:
        items["Salary-Commute"] = salary_earn/distance_travel
        commute.append(items)
        more_commute.append(items)



print("Total length of Salary-Commute:",len(commute))
print( )
print(commute)
```
